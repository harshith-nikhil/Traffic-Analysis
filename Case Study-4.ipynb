```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6,\
			7, 8, 9, 8, 9, 9, 8, 4, 4, 5, 4])
b = np.array([5, 4, 5, 6, 5, 8, 6, 7, 6, 7, \
			1, 2, 1, 2, 3, 2, 3, 9, 10, 9, 10])
c = np.array(list(zip(a, b))).reshape(len(a), 2)

plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Sample Dataset')
plt.scatter(a, b)
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(random_state=42)
visualise = KElbowVisualizer(model, k=(2, 7))
visualise.fit(c)
visualise.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
for i in range (2,5):
    model_sscore = KMeans(n_clusters=i)
    visualize_1 = SilhouetteVisualizer(model_sscore, colors='yellowbrick')
    visualize_1.fit(c)
    visualize_1.show()
```

### Using k=3 , as obtained from Elbow Method Result

```python
kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(c)
y_kmeans1 = kmeans1.predict(c)

plt.figure(figsize=(10, 6))
plt.scatter(c[:, 0], c[:, 1], c=y_kmeans1, s=50, cmap='viridis')

centers1 = kmeans1.cluster_centers_
plt.scatter(centers1[:, 0], centers1[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering with k=3')
plt.xlabel('Column 1')
plt.ylabel('Column 2')

plt.show()
```

### Using k=4, result as shown from Silhouette Score Method

```python
kmeans2 = KMeans(n_clusters=4)
kmeans2.fit(c)
y_kmeans2 = kmeans2.predict(c)

plt.figure(figsize=(10, 6))
plt.scatter(c[:, 0], c[:, 1], c=y_kmeans2, s=50, cmap='viridis')

centers2 = kmeans2.cluster_centers_
plt.scatter(centers2[:, 0], centers2[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering with k=4')
plt.xlabel('Column 1')
plt.ylabel('Column 2')

plt.show()
```
