```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import requests
import json
```

```python
nycrash_url='https://data.cityofnewyork.us/resource/f55k-p6yu.json'
response=requests.get(nycrash_url)
if response.status_code == 200:
    data = response.json()
    df_c = pd.json_normalize(data)
else:
    print(f"Failed to retrieve data: {response.status_code}")

df_c.head()
```

```python
df_c.info()
```

```python
def get_val_counts(coln):
    value_c=df_c[coln].value_counts()
    return value_c   
```

```python
df_c=df_c.drop(['contributing_factor_1','contributing_factor_2', 'ped_location', 'ped_action'], axis=1)
df_c.head()
```

```python
df_c.info()
```

```python
df_c['person_age']=df_c['person_age'].astype('float')
median_age=int(df_c['person_age'].median())
median_age
```

```python
df_c['safety_equipment']=df_c['safety_equipment'].fillna('Other')
df_c['emotional_status']=df_c['emotional_status'].fillna('Unknown')
df_c['bodily_injury']=df_c['bodily_injury'].fillna('Unknown')
df_c['ejection']=df_c['ejection'].fillna('Unknown')
df_c['position_in_vehicle']=df_c['position_in_vehicle'].fillna('Unknown')
df_c['complaint']=df_c['complaint'].fillna('Unknown')
df_c['person_sex']=df_c['person_sex'].fillna('U')
df_c['person_age']=df_c['person_age'].fillna(median_age)
```

```python
df_c=df_c.dropna()
```

```python
df_c['crash_date'] = df_c['crash_date'].str.split('T', expand=True)[0]
```

```python
df_c['unique_id']=df_c['unique_id'].astype('int')
df_c['collision_id']=df_c['collision_id'].astype('int')
df_c['vehicle_id']=df_c['vehicle_id'].astype('int')
df_c['crash_time'] = pd.to_datetime(df_c['crash_time'], format='%H:%M')
```

```python
df_c['crash_date'] = pd.to_datetime(df_c['crash_date'], format='%Y-%m-%d')
```
